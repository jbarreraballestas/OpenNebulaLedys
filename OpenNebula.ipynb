## Requisitos iniciales

Para la realización de la actividad previamente se realiza instalación de ubuntu 22.04

## Configurar repositorios
[Documentación repositorios](https://docs.opennebula.io/6.10/installation_and_configuration/frontend_installation/opennebula_repository_configuration.html#repositories)

Instalar paquetes necesarios

```bash
apt-get -y install gnupg wget apt-transport-https
```

Agregar llave de acceso a repositorios de OpenNebula

```bash
wget -q -O- https://downloads.opennebula.io/repo/repo2.key | gpg --dearmor --yes --output /etc/apt/keyrings/opennebula.gpg
```

Descargar y agregar el archivo de repositorio, actualizar repositorios
```bash
echo "deb [signed-by=/etc/apt/keyrings/opennebula.gpg] https://downloads.opennebula.io/repo/6.10/Ubuntu/22.04 stable opennebula" > /etc/apt/sources.list.d/opennebula.list
apt-get update
```
